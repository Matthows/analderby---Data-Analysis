```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import pingouin as pg
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
plt.style.use('ggplot')
```

# IMPORT CSV DATASET

```python
df_games = pd.read_csv('AssignmentData/games.csv')
df_club_games = pd.read_csv('AssignmentData/club_games.csv')
```

```python
df_games.columns
```

### Data Cleaning
Select relevant columns

```python
df_games = df_games[
['game_id', 'competition_id', 'season', 'round', 'date', 'home_club_id',
       'away_club_id', 'home_club_goals', 'away_club_goals',
       'home_club_position', 'away_club_position',
       'home_club_formation', 'away_club_formation', 'home_club_name',
       'away_club_name', 'competition_type']
]
```

```python
# Calculate Outcome (Win/Loss)
```

```python
def match_outcome(row):
    if row['home_club_goals'] > row['away_club_goals']:
        return 1
    elif row['home_club_goals'] < row['away_club_goals']:
        return -1
    else:
        return 0



# Apply the function to create a new column 'outcome'
df_games['outcome'] = df_games.apply(match_outcome, axis=1)
# Remove NaN value from 'home_club_formation' and 'away_club_formation' columns
df_games = df_games[~df_games['home_club_formation'].isna() & ~ df_games['away_club_formation'].isna()]
```

```python
# Create Dummy Variables for Formations
```

```python
pd.set_option('display.max_columns', None)


# Create dummy variables for 'home_club_formation'
formations_dummies = pd.get_dummies(df_games['home_club_formation'])

# Concatenate dummy variables with original DataFrame
df_with_dummies = pd.concat([df_games, formations_dummies], axis=1)
df_with_dummies
```

```python
#  Calculate Correlation with Match Outcome
```

```python
# Calculate correlation coefficients with 'outcome'
formation_correlations = df_with_dummies.iloc[:, -len(formations_dummies.columns):].corrwith(df_games['outcome'])

# Display correlations sorted by magnitude
formation_correlations_sorted = formation_correlations.sort_values(ascending=False)

print("Correlation of formations with match outcome:")
print(formation_correlations_sorted)
```

```python
# Select the top and bottom formations with highest positive and negative correlations
top_formations = formation_correlations_sorted.tail(10)  # Top 10 positive correlations
bottom_formations = formation_correlations_sorted.head(10)  # Top 10 negative correlations

# Plotting
plt.figure(figsize=(12, 6))

# Plot top positive correlations
plt.subplot(1, 2, 1)
top_formations.plot(kind='bar', color='b')
plt.title('Formations with Highest Positive Correlation')
plt.xlabel('Formation')
plt.ylabel('Correlation')

# Plot top negative correlations
plt.subplot(1, 2, 2)
bottom_formations.plot(kind='bar', color='r')
plt.title('Formations with Highest Negative Correlation')
plt.xlabel('Formation')
plt.ylabel('Correlation')

plt.tight_layout()
plt.show()
```
